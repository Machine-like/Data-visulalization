```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0, 10])
ypoints = np.array([5, 150])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([4,5])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = '*')
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:g')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints,marker = 'o', ms = 15)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'b')
plt.show()
```

```python
plt.plot(ypoints, '*--m',ms=20,mec='#4CAF50',mfc='hotpink')
```

```python
plt.plot(ypoints,linestyle = 'dotted')
```

```python
plt.plot(ypoints,linestyle = 'dashed')
```

```python
plt.plot(ypoints,color='r')
```

```python
plt.plot(ypoints,linewidth='12.5')
```

```python
y1 = np.array([3, 8, 5, 10])
y2 = np.array([6, 2, 11, 20])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([20,35,55,75])
y1 = np.array([25,60,57,28])
y2 = np.array([9,81,10,45])
x2 = np.array([85,65,42,22])
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
plt.grid()
plt.show()
```

```python
plt.grid(axis='x')
plt.show()
```

```python
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
#plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2
x = ([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```
